```python
import pandas as pd
import numpy as np
import tslearn, os, sys, datetime, pickle
import matplotlib.pyplot as plt
%matplotlib inline

from tslearn.clustering import TimeSeriesKMeans
from tslearn.utils import to_time_series_dataset
from tslearn.preprocessing import TimeSeriesResampler
```

```python
N_CLUSTERS = 10
DATA_DIR = "../data/"
RESULTS_DIR = "../results/"
```

```python
# All data imports
sales_train_validation = pd.read_csv(DATA_DIR+"sales_train_validation.csv")
```

```python
FEATURES = [i for i in sales_train_validation.columns if i.startswith("d_")]
```

```python
data = sales_train_validation[FEATURES].values.transpose().as
print(data.shape)
NUM_SAMPLES = len(data)
```

```python
series_data = to_time_series_dataset(data)
```

```python
clust_mod = TimeSeriesKMeans(**{
    "n_clusters":N_CLUSTERS,
    "max_iter":50,
    "tol":1e-06,
    "n_init":1,
    "metric":'euclidean', #"softdtw", "dtw"
    "max_iter_barycenter":100,
    "metric_params":None,
    "n_jobs":10,
    "dtw_inertia":False,
    "verbose":0,
    "random_state":None,
    "init":'k-means++',
})
```

```python
clust_mod.fit(series_data)
```

```python
np.unique(clust_mod.labels_, return_counts=1)
```

```python
NUM_ROWS = data.shape[0]
NUM_COLS = data.shape[1]
PLT_MAX, PLT_MIN = data.min(), data.max()
```

```python
for i in range(N_CLUSTERS):
    plt.plot(range(NUM_ROWS), data[:,[k for k,v in enumerate(clust_mod.labels_) if v == i]], alpha=0.9)
    plt.show()
```
