```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
import numpy as np
```

```python
NUM_DAYS = 128
data = np.arange(NUM_DAYS)
BS=8
NUM_TIMESTEPS = 28
```

```python
tsgX = TimeseriesGenerator(
    data=data,
    targets=data,
    length=NUM_TIMESTEPS,
    sampling_rate=1,
    stride=1,
    #start_index=0,
    end_index=NUM_DAYS - NUM_TIMESTEPS - 1,
    batch_size=BS
)

tsgY = TimeseriesGenerator(
    data=data,
    targets=data,
    length=NUM_TIMESTEPS,
    sampling_rate=1,
    stride=1,
    start_index=28,
    #end_index=NUM_DAYS - 1,
    batch_size=BS
)
```
